```python
import matplotlib.pyplot as plt 
```

```python
def y(x):
    return x**2

xval=[1,2,3,4,5]
y=[y(x) for x in xval]

plt.xticks(range(0, 11, 1))  # x축 0부터 10까지 1단위
plt.yticks(range(0, 27, 1))  # y축 0부터 100까지 1단위


plt.plot(xval,y,marker='o',color='r')
plt.xlim(0,10)
plt.ylim(0,27)
plt.show()
```
