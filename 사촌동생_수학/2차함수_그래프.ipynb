```python
import matplotlib.pyplot as plt 
import numpy as np

def y(x):
    return x**2

xval = np.arange(1,10,0.1)
y= [y(x) for x in xval]

plt.plot(xval,y,marker='o', color='r', markersize=3)

ax = plt.gca()

ax.set_xticks(range(0, 11, 1))  # x축 0부터 10까지 1단위
ax.set_yticks(range(0, 27, 1))  # y축 0부터 26까지 1단위

# 격자선 추가
plt.grid(True, which='both', axis='both', linestyle='--', linewidth=0.5)

plt.xlim(0,10)
plt.ylim(0,27)
plt.show()
```
